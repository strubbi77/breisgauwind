```python
Windkraft_je_ms_rel=pd.DataFrame()
Windkraft_je_ms_rel['m/s']=Windkraft_je_ms['m/s']
Windkraft_je_ms_rel['rel_V162-5.6']=Windkraft_je_ms['V162-5.6']/Windkraft_je_ms['V162-5.6'].max()
Windkraft_je_ms_rel['rel_V162-6.2']=Windkraft_je_ms['V162-6.2']/Windkraft_je_ms['V162-6.2'].max()
Windkraft_je_ms_rel['rel_V172-7.2']=Windkraft_je_ms['V172-7.2']/Windkraft_je_ms['V172-7.2'].max()
Windkraft_je_ms_rel['rel_V172-7.2SO']=Windkraft_je_ms['V172-7.2SO']/Windkraft_je_ms['V172-7.2SO'].max()
Windkraft_je_ms_rel['rel_EK100']=Windkraft_je_ms['EK100']/Windkraft_je_ms['EK100'].max()
Windkraft_je_ms_rel['rel_1998_600kw']=Windkraft_je_ms['1998_600kw']/Windkraft_je_ms['1998_600kw'].max()
Windkraft_je_ms_rel['rel_E-126']=Windkraft_je_ms['E-126']/Windkraft_je_ms['E-126'].max()
Windkraft_je_ms_rel['rel_Kitepower_Hawk']=Windkraft_je_ms['Kitepower_Hawk']/Windkraft_je_ms['Kitepower_Hawk'].max()
Windkraft_je_ms_rel['rel_Kitepower_Falcon']=Windkraft_je_ms['Kitepower_Falcon']/Windkraft_je_ms['Kitepower_Falcon'].max()
Windkraft_je_ms_rel['rel_SkySails_PN-14']=Windkraft_je_ms['SkySails_PN-14']/Windkraft_je_ms['SkySails_PN-14'].max()
Windkraft_je_ms_rel
```

```python
rel_max=Windkraft_je_ms[Windkraft_je_ms['m/s']==10]['max W/m²'].values[0]
Windkraft_je_ms_rel['gek max W/m² 10']=Windkraft_je_ms['max W/m²']/rel_max
rel_max=Windkraft_je_ms[Windkraft_je_ms['m/s']==15]['max W/m²'].values[0]
Windkraft_je_ms_rel['gek max W/m² 15']=Windkraft_je_ms['max W/m²']/rel_max
Windkraft_je_ms_rel
```
